```python
events = re.findall(r'[0-9]+\.json', events)
events = [x[:-5] for x in events]
```

```python
with open('b_classifier.pickle.dat', 'rb') as f:
    b_classifier = pickle.load(f)
```

```python
def prepr(data):
    df = data[0]
    current_teams = data[1]
    for side in ['start', 'end']:
        key_x = '{0}_location_x_normalized'.format(side)
        key_y = '{0}_location_y_normalized'.format(side)

        df[key_x] = df[key_x[:-11]] / max(df.start_location_x)
        df[key_y] = df[key_y[:-11]] / max(df.start_location_y)
        
    df.body_part = df.body_part.apply(lambda part: 0 if re.findall(r'Foot', part) else 1 if re.findall(r'Head', part) else 2)
    
    key_start_x = 'start_location_x_normalized'
    key_start_y = 'start_location_y_normalized'
    key_start_distance = 'start_distance'

    df[key_start_distance] = df.apply(lambda s: distance.euclidean((s[key_start_x], s[key_start_y]), goal), axis=1)
    return (df, current_teams)

def predicting(data):
    df = data[0]
    current_teams = data[1]
    res = {}
    fields = ['start_location_x_normalized', 'start_location_y_normalized', 'body_part', 'start_distance']
    df_test = df[fields]

    for i in range(df.index.size-1):
        key = '{0}'.format(df.player[i])
        data = pd.DataFrame([([df_test[df_test.columns[j]][i] for j in range(df_test.columns.size)])], columns=fields)
        val = round(b_classifier.predict_proba(data)[:, 1][0], 7)
        res[key] = (val, df.statsbomb_xg[i], df.start_location_x[i], df.start_location_y[i])
    return (res, current_teams)

def plotting(data):
    res = data[0]
    current_teams = data[1]
    x = [res[key][2] for key in res.keys()]
    y = [res[key][3] for key in res.keys()]

    #output_notebook()
    file = makename()
    output_file(file)

    source = ColumnDataSource(data=dict(
        x = x,
        y = y,
        player = [key for key in res.keys()],
        xg = [str(round(res[key][0], 2)) for key in res.keys()],
        sbxg = [str(round(res[key][1], 2)) for key in res.keys()],
        rad = [res[key][0]*10 for key in res.keys()],
        colors = ["#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(10*np.array(x), 10*np.array(y))],
    ))

    TOOLTIPS = [
        ("index", "$index"),
        ("player", "@player"),
        ("xg", "@xg"),
        ("sbxg", "@sbxg")
    ]
    p = figure(x_range=(0, 120), y_range=(0, 90), plot_width=120*5, plot_height=90*5, tooltips=TOOLTIPS,
               title="xG {0} : {1}".format(current_teams[0], current_teams[1]))

    p.circle('x', 'y', radius='rad', fill_color='colors', source=source) 

    save(p)
    return p

def makename():
    pattern = 'wc_xg'
    i = 0
    fnames = glob.glob('OUT/*.html')
    while 'OUT/{0}{1}.html'.format(pattern, i) in fnames:
        i += 1
    return 'OUT/{0}{1}.html'.format(pattern, i)

def ifWC(df):
    tm = list(set(list(df.team)))
    for team in tm:
        if team not in teams:
            return False, ()
    current_teams = (tm[0], tm[1])
    return True, current_teams
```

```python
pics = []

for event in tqdm.tqdm(events):
    BvsE = sb.Events(event_id=event)
    df_shot = BvsE.get_dataframe(event_type='shot')
    
    ifwc, current_teams = ifWC(df_shot)
    if not ifwc:
        continue
        
    funcs = [(df_shot, current_teams), prepr, predicting, plotting]
    
    pics.append(reduce(lambda a, b: b(a), funcs))
    
```
