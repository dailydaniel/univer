```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pprint as pp

from l5 import readata, createMatrix, makea, makeArr, frange, qd, lagr
```

```python
path = 'l5.txt'
x, y = readata(path)
print('x: {0}\ny: {1}'.format(x, y))
```

```python
A = []

for n in [2, 3, 4, 11]:
    m = createMatrix(x, y, n)
    A.append(makea(m))
#     pp.pprint(m)
#     print()
    
for a in A:
    pp.pprint('коэффициенты для многочлена {0}-й степени = {1}'.format(len(a)-1, a))
```

```python
delta = 0.005
X = frange(x[0], x[-1], delta)

fig, ax = plt.subplots(figsize=(5, 8))
for a in A:
    ax.plot(X, [makeArr(a, el) for el in X], label=str(len(a)-1))
ax.set_title('x')
ax.legend(loc='upper left')
ax.set_ylabel('y')
ax.set_xlim(xmin=-1, xmax=4)
ax.set_ylim(ymin=-1, ymax=7)
fig.tight_layout()
 
plt.show()
```

```python
for a in A:
    pp.pprint('квадратичное отклонение для многочлена {0}-й степени = {1}'.format(len(a)-1, qd(x, y, x, y, a)))
```

```python
l_x = [x[i] for i in range(len(x)) if i in (1, 3, 5)]
l_y = [y[i] for i in range(len(y)) if i in (1, 3, 5)]

pp.pprint('квадратичное отклонение для многочлена лагранжа = {0}'.format(qd(l_x, l_y, x, y, None, lagr)))
```
