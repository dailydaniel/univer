### Задача:

Решить задачу Коши методом Рунге-Кутты со значениями шага $h_1 = 0.1$ и $h_2=0.01$ и сравнить результаты графически.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

### Начальные условия:

$y^1 = \frac{y}{x} + x^2sinx$, 

$y(1)=1.3011687$, $x \in [1,2]$

```python
y_0 = 1.3011687
x_0 = 1.
x_1 = 3
h_1 = 0.1
h_2 = 0.01
h_3 = 2.
def f(x,y):
    return y/x + (x**2)*np.sin(x)
```

```python
def Runge_Kutta(x,y,x_1,h):
    x_arr = [x]
    y_arr = [y]
    while x_1 > x:
        K1 = h*f(x,y)
        K2 = h*f(x+0.5*h,y+0.5*K1)
        K3 = h*f(x+0.5*h,y+0.5*K2)
        K4 = h*f(x+h,y+K3)
        y += 1/6*(K1+2*K2+2*K3+K4)
        x += h
        x = round(x,2)
        x_arr += [x]
        y_arr += [y]
    return {'x': np.array(x_arr), 'y': np.array(y_arr)}
```

```python
plt.style.use('seaborn')
data1 = Runge_Kutta(x_0,y_0,x_1,0.5)
data2 = Runge_Kutta(x_0,y_0,x_1,h_2)
fig = plt.figure(figsize = (12.5,7))
ax = fig.add_subplot(111)
ax.plot(data1['x'],data1['y'],'-g', label = '$h_1$='+str(h_1),linewidth = 2)
ax.plot(data2['x'],data2['y'],'--k', label = '$h_2$='+str(h_2),linewidth = 2)
plt.legend(fontsize = 13)
plt.title('Runge–Kutta methods', size = 15)
plt.show()
```

```python
print('Максимальная разница:',np.max(np.abs(data1['y']-data2['y'][::10])))
print('Индекс с максимаьлной разницей:',np.where(np.abs(data1['y']-data2['y'][::10]) == np.max(np.abs(data1['y']-data2['y'][::10])))[0][0])
```

```python
plt.style.use('seaborn')
data1 = Runge_Kutta(x_0,y_0,150,h_1)
data2 = Runge_Kutta(x_0,y_0,150,h_2)
data3 = Runge_Kutta(x_0,y_0,150,h_3)
fig = plt.figure(figsize = (12.5,7))
ax = fig.add_subplot(111)
ax.plot(data1['x'],data1['y'],'-g', label = '$h_1$='+str(h_1),linewidth = 2)
ax.plot(data2['x'],data2['y'],'--k', label = '$h_2$='+str(h_2),linewidth = 2)
ax.plot(data3['x'],data3['y'],'--r', label = '$h_3$='+str(h_3),linewidth = 2)
plt.legend(fontsize = 13)
plt.title('Runge–Kutta methods', size = 15)
plt.show()
#print('при h=0.1 x = ',data1['x'])
#print('при h=0.1 y = ',np.round(data1['y'],15))
#print('\n\nпри h=0.01 x = ',data2['x'])
#print('при h=0.01 y = ',np.round(data2['y'],15))
```
