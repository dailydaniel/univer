```python
import pandas as pd
import string
import numpy as np
import re

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

from tqdm import tqdm

import matplotlib.pyplot as plt

from collections import defaultdict

from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('../All_irregular_verb_list.csv', sep=';', encoding='utf-8')
df = df[['First', 'Second', 'Third']]
```

```python
for column in df.columns[1:]:
    df[column] = df[column].apply(lambda x: re.sub(r'[/.*]', '', x))
    df[column] = df[column].apply(lambda x: re.sub(r'[^a-zA-Z]', '', x))

criteria = df['First'].map(lambda x: len(set(x) & set(string.ascii_uppercase)) == 0)
df = df[criteria]
df = df.reset_index(drop=True)
```

```python
char_emb = {char: i+1 for i, char in enumerate(string.ascii_lowercase)}
char_emb['0'] = 0
```

```python
decoder = {val: key for key, val in char_emb.items()}

def decode(idx, decoder):
    t = ''.join([decoder[ix] for ix in idx])
    return re.sub(r'[0.*]', '', t)

def line_decode(idx, decoder):
    t = ''.join([decoder[ix] for ix in idx])
    if t.find('0') == -1:
        t = t[:int(len(t)/2)] + ' ' + t[int(len(t)/2):]
        return t
    t = re.sub(r'0*$', '', t)
    t = re.sub(r'\d(0*)\d', ' ', t)
    return t
```

```python
max_len = 0
for column_id in range(len(df.columns)):
    for i in range(df.index.size-1):
        if len(df[df.columns[column_id]][i]) > max_len:
            max_len = len(df[df.columns[column_id]][i])
```

```python
def word2emb(emb, n, word):
    res = np.zeros(n, dtype=int)
    for i, char in enumerate(word):
        res[i] = emb[char]
    return res

for column in df.columns:
    print(column)
    df[column] = df[column].apply(lambda word: word2emb(char_emb, max_len, word))
```

```python
data, labels_sec_form, labels_th_form = df['First'], df['Second'], df['Third']
data, labels_sec_form, labels_th_form = data.map(lambda x: torch.from_numpy(x)), labels_sec_form.map(lambda x: torch.from_numpy(x)), labels_th_form.map(lambda x: torch.from_numpy(x))
```

```python
def acc(write_res = False):
    accuracy = []
    for sentence, target_s, targets_th in zip(data, labels_sec_form, labels_th_form):
        with torch.no_grad():
            tag_scores = model(sentence)
            _, preds = torch.max(tag_scores, 1)
            first = line_decode(sentence.tolist(), decoder)
            gt = line_decode(torch.cat((target_s, targets_th)).tolist(), decoder)
            pred = line_decode(preds.tolist(), decoder)
            accuracy.append(int(gt == pred))
            if write_res:
                print('{0} => gt: {1}, pred: {2}'.format(first, 
                                                         gt, 
                                                         pred))
    return sum(accuracy) / len(accuracy)
```

```python
EMBEDDING_DIM = 8
HIDDEN_DIM = 8
```

```python
class LSTMVerbFormer(nn.Module):

    def __init__(self, embedding_dim, hidden_dim, vocab_size):
        super(LSTMVerbFormer, self).__init__()
        self.hidden_dim = hidden_dim

        self.char_embeddings = nn.Embedding(vocab_size, embedding_dim)

        self.lstm_sec_form = nn.LSTM(embedding_dim, hidden_dim, bidirectional=True) # False
        
        self.lstm_th_form = nn.LSTM(embedding_dim, hidden_dim, bidirectional=True) # False

        self.hidden2tag = nn.Linear(hidden_dim * 2, vocab_size * 2) # without * 2

    def forward(self, sentence):
        embeds = self.char_embeddings(sentence)
        lstm_out_sec_form, _ = self.lstm_sec_form(embeds.view(len(sentence), 1, -1))
        lstm_out_th_form, _ = self.lstm_th_form(embeds.view(len(sentence), 1, -1))
        comb = torch.cat((lstm_out_sec_form, lstm_out_th_form))
        tag_space = self.hidden2tag(comb.view(len(sentence) * 2, -1))
        tag_scores = F.log_softmax(tag_space, dim=1)
        return tag_scores
```

```python
model = LSTMVerbFormer(EMBEDDING_DIM, HIDDEN_DIM, len(char_emb))
loss_function = nn.NLLLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
avg_loss = []
accuracy = []

for epoch in tqdm(range(400)): 
    
    running_loss = []
    for sentence, targets_s, targets_th in zip(data, labels_sec_form, labels_th_form):
        model.zero_grad()

        tag_scores = model(sentence)
        targets = torch.cat((targets_s, targets_th))
    
        loss = loss_function(tag_scores, targets)
        loss.backward()
        optimizer.step()
        
        running_loss.append(loss.item())
    avg_loss.append(round(sum(running_loss) / len(running_loss), 4))
    accuracy.append(round(acc(), 4))
```

```python
X, Y = range(len(accuracy)), accuracy

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(X, Y, label='lstm')
ax.set_title('n_epoch')
ax.legend(loc='upper left')
ax.set_ylabel('accuracy')
fig.tight_layout()
 
plt.show()
```

```python
X, Y = range(len(avg_loss)), avg_loss

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(X, Y, label='lstm')
ax.set_title('n_epoch')
ax.legend(loc='upper left')
ax.set_ylabel('loss')
fig.tight_layout()
 
plt.show()
```

```python
embedding_by_avg = {}

for sent in tqdm(data):
    with torch.no_grad():
        t = model.char_embeddings(sent)
        t_sec, _ = model.lstm_sec_form(t.view(len(sent), 1, -1))
        t_sec = t_sec.view(len(sent), -1)
        t_th, _ = model.lstm_th_form(t.view(len(sent), 1, -1))
        t_th = t_th.view(len(sent), -1)      
        t = torch.cat((t_sec, t_th))
    
        tag_scores = model(sent)
        _, preds = torch.max(tag_scores, 1)
    pred = line_decode(preds.tolist(), decoder)
    try:
        embedding_by_avg['{0} - {1} - {2}'.format(decode(sent.tolist(), decoder), pred.split()[0], pred.split()[1])] = (sum(t) / t.shape[0]).numpy()
    except:
        print(sent, pred)
```

```python
keys = list(embedding_by_avg.keys())
vectors = [embedding_by_avg[key] for key in keys]
```

```python
n = 20
best_n = n
s_score = [0]
best_clusterization = None
for k_n in tqdm(range(n, 50)):
    kmeans = KMeans(n_clusters=k_n)
    kmeans.fit(vectors)
    current_s_score = silhouette_score(vectors, kmeans.labels_)
    if current_s_score > max(s_score):
        best_n = k_n
        best_clusterization = kmeans
    s_score.append(current_s_score)
s_score = s_score[1:]
best_n
```

```python
X, Y = range(n, n + len(s_score)), s_score

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(X, Y, label='silhouette_score')
ax.set_title('n_clusters')
ax.legend(loc='upper left')
ax.set_ylabel('score')
fig.tight_layout()
 
plt.show()
```

```python
res = defaultdict(list)
for label, key in zip(best_clusterization.labels_, keys): # kmeans
    res[label].append(key)
```

```python
for cluster, val in res.items():
    print('cluster: {0}\n'.format(cluster))
    for el in val:
        print(el)
    print('\n', '='*50, '\n')
```
